```python
import pandas as pd
```

```python
df = pd.read_csv('chicago.csv').dropna(how='all')
df.head()
```

```python
df.info()
```

```python
df['Employee Annual Salary'] = df['Employee Annual Salary'].str.replace('$', '').astype('float')
```

```python
df['Department'] = df['Department'].astype('category')
```

```python
df.info()
```

```python
mask = df['Position Title'].str.lower().str.contains('civil engineer')
df[mask]
```

```python
mask = df['Position Title'].str.lower().str.startswith('civil')
df[mask]
```

```python
df['Name'] = df['Name'].str.strip()
```

```python
df = pd.read_csv('chicago.csv', index_col='Name').dropna(how='all')
df.head()
```

```python
df.index
```

```python
df.index = df.index.str.strip().str.title()
```

```python
df.columns = df.columns.str.upper()
```
