```python
import pandas as pd
```

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
pd.concat([week1, week2], ignore_index=True)
```

```python
sales = pd.concat([week1, week2], keys=['week1', 'week2'])
```

```python
sales.loc[('week1', 249)]
```

```python
week2.append(week1, ignore_index=True)
```

# Merge

```python
week1.merge(week2, how='inner', on='Customer ID', suffixes=(' - Week 1', ' - Week 2'))
```

```python
week1.merge(week2, how='inner', on=['Customer ID', 'Food ID'])
```

```python
week1.merge(
            week2, 
            how='outer', 
            on='Customer ID', 
            suffixes=(' - Week 1', ' - Week 2'),
            indicator=True
            )
```

```python
week1.merge(foods, how='left', on="Food ID")
```

```python
week2.merge(customers, how='left', left_on='Customer ID', right_on='ID', sort=True)
```
