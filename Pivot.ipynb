```python
df = pd.read_csv('foods.csv')
```

```python
df
```

```python
import numpy as np
df.pivot_table(values='Spend', index='Item', aggfunc=np.sum)
```

```python
df.pivot_table(values='Spend', index='Gender', aggfunc=np.sum)
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc="count")
```

```python
df.pivot_table(
    values='Spend', 
    index=['Gender', 'Item'], 
    columns='City', 
    aggfunc=np.sum
)
```
